```python
df = pd.read_csv('kc_house_data.csv')
```

```python
for col in ['date', 'bedrooms', 'bathrooms', 'floors', 'waterfront', 'view',
            'condition', 'grade', 'yr_built', 'yr_renovated', 'zipcode']:
    print(col, '\n', df[col].value_counts(normalize=True).head(), '\n\n')
```

```python
df.describe()
```

```python
df2 = df[['date', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
          'condition', 'grade', 'yr_built', 'zipcode', 'price']]

scatter = pd.plotting.scatter_matrix(df2, figsize=(20,20))
```

```python
df['price_per_sqft_lot'] = df['price']/df['sqft_lot']
df['price_per_sqft_living'] = df['price']/df['sqft_living']

geo1 = df.loc[:, ['lat','long','price_per_sqft_lot']]
geo2 = df.loc[:, ['lat','long','price_per_sqft_living']]
geo4 = df.loc[:, ['zipcode','price_per_sqft_living','price_per_sqft_lot']]
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.boxplot(geo1.price_per_sqft_lot)
ax1.set_title('price_per_sqft_lot')

ax2.boxplot(geo2.price_per_sqft_living)
ax2.set_title('price_per_sqft_living')
```
